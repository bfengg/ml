```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/synthetic_commitment_data.csv')
```

```python
df['actual_compute_spend'] = (
    df['six_month_avg'] * 1.05 + 
    df['forecasted_growth'] * 1000 +  
    df['utilization_rate'] * 500 +  
    (1 - df['stability_factor']) * 2000  
)

# Clip the spend to avoid negative or extreme values
df['actual_compute_spend'] = np.clip(df['actual_compute_spend'], 20000, 200000)
df.drop(columns=['one_month_std', 'forecasted_growth', 'six_month_std', 'stability_factor', 'commitment_percentage'], inplace=True)
df.to_csv('../data/synthetic_compute_spend.csv', index=False)
df
```
